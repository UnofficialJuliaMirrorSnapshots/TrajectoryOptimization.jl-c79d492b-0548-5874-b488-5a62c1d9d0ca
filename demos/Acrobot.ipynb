```julia
using TrajectoryOptimization
using LinearAlgebra
using Plots
using MeshCat
using RigidBodyDynamics
using MeshCatMechanisms
```

# Define dynamics
Dynamics should be defined using in-place operations

```julia
function acrobot_dynamics!(Ẋ, X, U)
    m1 = 1.0; #mass of link 1
    m2 = 1.0; #mass of link 2
    L1 = 1.0; #length of link 1
    L2 = 1.0; #length of link 2
    Lc1 = 0.5; #distance from pivot to center of link 1
    Lc2 = 0.5; #distance from pivot to center of link 2
    J1 = (1.0/3.0)*m1*L1*L1; #moment of inertia of link 1 about pivot
    J2 = (1.0/3.0)*m2*L2*L2; #moment of inertia of link 2 about pivot
    g = 9.81;

    q = X[1:2];
    qd = X[3:4];

    if isfinite(q[2])
        s = sin(q[2])
        c = cos(q[2])
    else
        s = Inf
        c = Inf
    end

    s1 = sin(q[1]);
    s12 = sin(q[1]+q[2]);
    s2 = sin(q[2]);
    c2 = cos(q[2]);
    
    H = [J1+J2+m2*L1*L1+2*m2*L1*Lc2*c2  J2+m2*L1*Lc2*c2;  J2+m2*L1*Lc2*c2  J2];
    C = [-2*m2*L1*Lc2*s2*qd[2]  -m2*L1*Lc2*s2*qd[2];  m2*L1*Lc2*s2*qd[1]  0];
    G = [(m1*Lc1 + m2*L1)*g*s1+m2*g*L2*s12;  m2*g*L2*s12];
    B = [0; 1];

    qdd = -H\(C*qd + G - B*U');

    Ẋ[1:2] = qd
    Ẋ[3:4] = qdd
    return nothing
end
```

# Create a model

```julia
n = 4
m = 1

#
#model = Model(acrobot_dynamics!,n,m)

traj_folder = joinpath(dirname(pathof(TrajectoryOptimization)),"..");
urdf_folder = joinpath(traj_folder, "dynamics/urdf");
urdf = joinpath(urdf_folder, "acrobot.urdf");
model = Model(urdf,[0.;1.]); # underactuated, only control for second "elbow" joint
```

# Create an objective

```julia
x0 = [0.0;0.0;0.0;0.0]
xf = [pi;0.0;0.0;0.0]

Q = 1.0*Diagonal(I,n)
Qf = 1000.0*Diagonal(I,n)
R = 0.1*Diagonal(I,m)

dt = 0.05
tf = 10.0
N = tf/dt+1
t = dt*collect(0:(N-1))

obj_uncon = LQRObjective(Q, R, Qf, tf, x0, xf)

#u_min = -10.
#u_max = 10.

#x_min = [-Inf;-Inf;-Inf;-Inf]
#x_max = [Inf; Inf; Inf; Inf]
#obj_con = ConstrainedObjective(obj_uncon,u_min=u_min, u_max=u_max, x_min=x_min, x_max=x_max);
```

# Create a solver
change default options as necessary

```julia
opts = SolverOptions()
opts.verbose = true;
opts.square_root = true;

solver = Solver(model,obj_uncon,dt=dt,integration=:rk3,opts=opts);
```

# Provide initial trajectory
either: U; X,U; none

```julia
U0 = randn((solver.model.m,solver_uncon.N-1));
```

# Solve problem

```julia
results, stats = solve(solver_uncon,U0);
```

```julia
xhist = hcat(results.X...)
plot(xhist[1:2,:]',title="Acrobot",xlabel="Time Step",ylabel="Theta",label=["Theta1","Theta2"])
```
